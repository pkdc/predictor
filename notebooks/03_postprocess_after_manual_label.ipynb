```python
emails_df.columns
```

```python
emails_df = emails_df.loc[:, ~emails_df.columns.str.startswith('Unnamed:')]
emails_df.rename(columns={'DISC_fiNl': 'DISC_final'}, inplace=True)
emails_df.head()
```

```python
emails_df = emails_df[~(emails_df['DISC_manual'] == 'N')]
emails_df.head()
```

```python
emails_df.isnull().sum()
```

```python
disc_order = ['D', 'I', 'S', 'C']

def parse_disc_manual(text):
    # print(text)
    # If already a list, return it as-is (but sorted)
    if isinstance(text, list):
        return sorted([x for x in text if x in disc_order], key=lambda x: disc_order.index(x))

    # If NaN or not string, return empty list
    if pd.isna(text) or not isinstance(text, str):
        return []

    # Parse from string: "(D,I)" or "['D','S']"
    parts = text.strip("()[]").split(',')
    labels = [p.strip() for p in parts if p.strip() in disc_order]

    return sorted(labels, key=lambda x: disc_order.index(x))

emails_df['DISC_manual'] = emails_df['DISC_manual'].apply(parse_disc_manual)
emails_df.head()
```

```python
emails_df = emails_df[emails_df['Entire-Message'].notna()]
emails_df['DISC_rule'] = pd.NA
emails_df['DISC_final'] = pd.NA
emails_df = emails_df.reset_index(drop=True)
emails_df.info()
```
